```python
import pandas as pd

# Assuming your dataset is stored in a JSON file named 'dataset.json'
dataset_path = 'gdrive/My Drive/mediqa-m3-clinicalnlp2024/valid_inputonly.json'

# Read the dataset into a DataFrame
df = pd.read_json(dataset_path)

# Select desired columns
selected_columns = ['encounter_id', 'image_ids', 'query_title_en', 'query_content_en']
selected_df = df[selected_columns]

# Print the selected DataFrame
print(selected_df)
```

```python
import os
import pickle
import pandas as pd
from skimage.io import imread
from skimage.transform import resize

# Load the selected DataFrame containing only the desired columns
# Assuming you already have selected_df DataFrame
#selected_df = pd.read_json('selected_dataframe.json')  # Replace 'selected_dataframe.json' with the actual path to your selected DataFrame

# Load the saved SVM model from the file
model_file_path = 'gdrive/My Drive/svm_modelM3G2.pkl'
with open(model_file_path, 'rb') as file:
    loaded_model = pickle.load(file)

# Function to predict classes for images in a list
def predict_classes_for_images(image_paths):
    predicted_classes = []
    for image_path in image_paths:
        try:
            image = imread(image_path)
            image = resize(image, (100, 100))  # Resize image to match the size used during training
            flattened_image = image.flatten().reshape(1, -1)  # Reshape to 1 sample
            predicted_class = loaded_model.predict(flattened_image)
            predicted_classes.append(predicted_class[0])
        except Exception as e:
            print(f"Error loading or processing image {image_path}: {e}")
            predicted_classes.append(None)
    return predicted_classes

# Iterate over each row of the DataFrame
# Iterate over each row of the DataFrame
output_list = []
for index, row in selected_df.iterrows():
    encounter_id = row['encounter_id']
    image_paths = [os.path.join('gdrive/My Drive/images_final/images_train/', image) for image in row['image_ids']]
    predicted_classes = predict_classes_for_images(image_paths)

    # Remove duplicates while preserving order
    unique_predicted_classes = []
    for pred_class in predicted_classes:
        if pred_class not in unique_predicted_classes:
            unique_predicted_classes.append(pred_class)

    output = ' '.join([str(pred) for pred in unique_predicted_classes])
    output_list.append(output)

# Add the output list to the DataFrame
selected_df['predicted_classes'] = output_list

# Save the modified DataFrame
selected_df.to_json('predicted_classes_dataframe.json', orient='records')
```

```python
import pandas as pd
from pyqna.models.reading_comprehension.transformer_models import TransformerQnAModel

# Load the selected DataFrame containing only the desired columns
# Assuming you already have selected_df DataFrame
#selected_df = pd.read_json('gdrive/My Drive/predicted_classes_dataframe.json')  # Replace 'selected_dataframe.json' with the actual path to your selected DataFrame

# Function to combine context with predicted classes and English queries
def combine_context_with_predicted_classes_and_queries(context, predicted_classes, query_title):
    combined_context = f"Expected disease name is {' '.join(predicted_classes)} {context}"
    query_treatment = "What is the treatment of the disease?"
    query_disease_name = f"What is the disease name? {query_title}"
    return combined_context, [query_treatment, query_disease_name]

# Function to execute queries using the instantiated model
def execute_queries(context, queries, model):
    answers = model.get_answer(context, queries)
    return [answer['answer'] for answer in answers]

# Instantiate the models
model_configs = [
    #{"model_name": "Chetna19/bart-large-finetuned-squadv1_qa_model", "pre_trained": True},
    #{"model_name": "hagara/biobert-qa", "pre_trained": True},
    #{"model_name": "armageddon/electra-base-squad2-covid-qa-deepset", "pre_trained": True},
    {"model_name": "longluu/Medical-QA-deberta-MRQA-COVID-QA", "pre_trained": True}
]

models = [TransformerQnAModel(config) for config in model_configs]

# Iterate over each row of the DataFrame
output_list = []
for index, row in selected_df.iterrows():
    context = row['query_content_en']
    predicted_classes = row['predicted_classes'].split()  # Assuming predicted_classes is a string with space-separated classes
    query_title = row['query_title_en']
    combined_context, queries = combine_context_with_predicted_classes_and_queries(context, predicted_classes, query_title)

    model_outputs = []
    for model, query in zip(models, queries):
        # Truncate the context if needed to match the maximum input length
        max_input_length = 512  # or any other suitable value
        combined_context = combined_context[:max_input_length]
        answers = execute_queries(combined_context, [query], model)
        model_outputs.append(answers[0])

    output_list.append(model_outputs)

# Add the output list to the DataFrame
selected_df['model_outputs3'] = output_list

# Save the modified DataFrame
selected_df.to_json('gdrive/My Drive/model_outputs_dataframe.json', orient='records')
```

```python
import torch
import clip
from PIL import Image
import pandas as pd

device = "cuda" if torch.cuda.is_available() else "cpu"
model, preprocess = clip.load("ViT-B/32", device=device)
def truncate_text_list(text_list, max_length):
    truncated_list = []
    for text in text_list:
        # Truncate the text if it exceeds the maximum length
        truncated_text = text[:max_length]
        truncated_list.append(truncated_text)
    return truncated_list
# Function to process one image and compute probabilities
# Function to process one image and compute probabilities
def process_image(image_path, text):
    print(image_path, text)
    image = preprocess(Image.open(image_path)).unsqueeze(0).to(device)
    text_token = clip.tokenize([text]).to(device)

    with torch.no_grad():
        image_features = model.encode_image(image)
        text_features = model.encode_text(text_token)
        logits_per_image, logits_per_text = model(image, text_token)
        probs = logits_per_image.softmax(dim=-1).cpu().numpy()

    return probs


df = df2

# Initialize an empty list to store the results
results = []

# Iterate through each row of the DataFrame
for index, row in df.iterrows():
    encounter_id = row['encounter_id']
    image_id = row['image_ids']
    combined_output = row['predicted_classes']
    # Pick the first image from the list
    image_id = image_id[0]
    image_path = f"gdrive/My Drive/images_final/images_train/{image_id}"  # Replace 'path_to_images' with the actual path to your images

    # Process the image and compute probabilities
    probabilities = process_image(image_path, truncate_text_list(combined_output, max_length=77))

    # Get the index of the maximum probability
    max_prob_index = probabilities.argmax()

    # Print the corresponding label using the index
    max_prob_label = combined_output[max_prob_index]  # Replace 'your_label_list' with your own list of labels or class names
    print(max_prob_label)
    # Append the result to the list
    results.append({'encounter_id': encounter_id, 'max_prob_label': max_prob_label})

# Create a new DataFrame from the results list
result_df = pd.DataFrame(results)

# Print the result DataFrame
print(result_df)
```

```python
from googletrans import Translator

# Function to translate disease names
def translate_disease_names(disease_names, src_lang, dest_lang):
    translator = Translator()
    translations = [translator.translate(name, src=src_lang, dest=dest_lang).text for name in disease_names]
    return translations

# Translate disease names to Chinese (Simplified)
chinese_translations = translate_disease_names(df['predicted_classes'], 'en', 'zh-cn')
df['max_prob_label_chinese'] = chinese_translations

# Translate disease names to Spanish
spanish_translations = translate_disease_names(df['predicted_classes'], 'en', 'es')
df['max_prob_label_spanish'] = spanish_translations
```
